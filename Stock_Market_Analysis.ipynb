```python
# 5a. Ensure 'stock' is a DataFrame by downloading the data properly
stock = yf.download("RELIANCE.NS", start="2020-01-01", end="2024-01-01")

# 5b. Verify it worked (Should print: <class 'pandas.core.frame.DataFrame'>)
print(type(stock))

# 5c. Plotting the Close Price
plt.figure(figsize=(12,6))
plt.plot(stock['Close'], color='blue', linewidth=1.5)

# Adding details to the graph
plt.title("Reliance Industries Stock Price Trend (2020 - 2024)", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Price (INR)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calculating 50-day and 200-day moving averages
stock['MA50'] = stock['Close'].rolling(window=50).mean()
stock['MA200'] = stock['Close'].rolling(window=200).mean()

# Plotting the Close price vs Moving Averages
plt.figure(figsize=(14,7))
plt.plot(stock['Close'], label='Close Price', alpha=0.5)
plt.plot(stock['MA50'], label='50 Day MA', color='orange')
plt.plot(stock['MA200'], label='200 Day MA', color='red')

plt.title("Reliance Industries: Moving Average Analysis")
plt.xlabel("Date")
plt.ylabel("Price (INR)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate percentage change
stock['Daily Return'] = stock['Close'].pct_change()

# Display the first few rows to see the new column
stock[['Close', 'Daily Return']].head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(stock['Daily Return'].dropna(), bins=100, color='purple', kde=True)
plt.title("Distribution of Daily Returns - Reliance")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(14,5))

# Use .values.flatten() to convert the 2D column into a 1D array
volume_data = stock['Volume'].values.flatten()

plt.fill_between(stock.index, volume_data, color="skyblue", alpha=0.4)
plt.plot(stock.index, volume_data, color="Slateblue", alpha=0.6)

plt.title("Reliance Daily Trading Volume")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

Project Insights: Reliance Industries (2020-2024)
Trend Identification: The 50-day Moving Average stayed consistently above the 200-day MA for most of the period, indicating a strong bullish trend.

Volatility: The Daily Return distribution is centered around 0, with "fat tails" suggesting occasional high-volatility events (common in major equity stocks).

Liquidity: Trading volume shows significant spikes during specific earnings periods or market news, showing high liquidity for the stock.

```python
# Download Nifty 50 data as the benchmark
nifty = yf.download("^NSEI", start="2020-01-01", end="2024-01-01")

# Collapse MultiIndex if necessary
for df in [stock, nifty]:
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

# Calculate Cumulative Returns
stock['Cumulative Return'] = (1 + stock['Close'].pct_change()).cumprod()
nifty['Cumulative Return'] = (1 + nifty['Close'].pct_change()).cumprod()

# Plot the comparison
plt.figure(figsize=(12,6))
plt.plot(stock['Cumulative Return'], label='Reliance', color='blue')
plt.plot(nifty['Cumulative Return'], label='Nifty 50 (Market)', color='black', linestyle='--')
plt.title("Cumulative Returns: Reliance vs. Nifty 50 Index")
plt.ylabel("Growth of 1 INR Investment")
plt.legend()
plt.show()
```

```python
# Assuming 5% annual risk-free rate (standard for India T-Bills)
risk_free_rate = 0.05 
daily_rf = (1 + risk_free_rate)**(1/252) - 1

# Calculate excess returns
excess_returns = stock['Daily Return'].dropna() - daily_rf

# Annualized Sharpe Ratio
sharpe_ratio = (excess_returns.mean() / excess_returns.std()) * np.sqrt(252)

print(f"Annualized Sharpe Ratio for Reliance: {sharpe_ratio:.2f}")
```

```python
df_ml = stock[['Close']].copy()
if isinstance(df_ml.columns, pd.MultiIndex):
    df_ml.columns = df_ml.columns.get_level_values(0)

# Create features: prices of the previous 5 days
for i in range(1, 6):
    df_ml[f'Prev_Day_{i}'] = df_ml['Close'].shift(i)

# Drop rows with NaN values created by shifting
df_ml.dropna(inplace=True)

# Define Features (X) and Target (y)
X = df_ml[['Prev_Day_1', 'Prev_Day_2', 'Prev_Day_3', 'Prev_Day_4', 'Prev_Day_5']]
y = df_ml['Close']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Split into 80% training and 20% testing
# We don't shuffle because stock data is a time series (order matters!)
split = int(0.8 * len(df_ml))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Initialize and Train the Model
model = LinearRegression()
model.fit(X_train, y_train)

# Make Predictions
predictions = model.predict(X_test)
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test.index, y_test, label="Actual Price", color='blue', alpha=0.7)
plt.plot(y_test.index, predictions, label="Predicted Price", color='red', linestyle='--')
plt.title("Reliance Price Prediction: Actual vs Predicted")
plt.xlabel("Date")
plt.ylabel("Price (INR)")
plt.legend()
plt.show()

# Print Accuracy Metrics
print(f"R-squared Score: {r2_score(y_test, predictions):.4f}")
```
